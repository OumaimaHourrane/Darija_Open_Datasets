```python
import csv
from itertools import zip_longest
title1=[]
location1=[]
distance1=[]
description1=[]
verdict1=[]
rating1=[]
experience1=[]
```

```python
def parse(url):
    options=webdriver.ChromeOptions()
    #options.add_arguments('/content/chromedrvier.exe')
    options.add_argument('--incognito')
    driver=webdriver.Chrome(options=options)
    driver.get(url)
    src=driver.page_source
    soup=BeautifulSoup(src,'html.parser')
    lists=soup.find_all('div',{'class':'d20f4628d0'})
    
    
    
    for list_ in lists:
        
        try:
            title=list_.find('div',{'class':'fcab3ed991 a23c043802'}).text
        except AttributeError as e:
            title=""
        title1.append(title)
        try:
            location=list_.find('span',{'class':'f4bd0794db b4273d69aa'}).text
        except AttributeError as e:
            location=""
        location1.append(location)
        try:
            distance=list_.find('span',attrs={'aria-expanded':'false','data-testid':'distance'}).text
        except AttributeError as e:
            distance=""
        distance1.append(distance)
        try:
            description=list_.find('div',{'class':'d8eab2cf7f'}).text
        except AttributeError as e:
            description=""
        description1.append(description)
        try:
            verdict=list_.find('div',{'class':'b5cd09854e f0d4d6a2f5 e46e88563a'}).text
        except AttributeError as e:
            verdict=""
        verdict1.append(verdict)
        try:
            rating=list_.find('div',{'class':'d10a6220b4'}).text
        except AttributeError as e:
            rating=""
        rating1.append(rating)
        try:
            experience=list_.find('div',{'class':'d8eab2cf7f c90c0a70d3 db63693c62'}).text
        except AttributeError as e:
            experience=""
        experience1.append(experience)
        
    info=[title1,location1,distance1,description1,verdict1,rating1,experience1]
    exported=zip_longest(*info)
    with open("C:/Users/Soukaina/OneDrive/Bureau/Hotels.csv","w",newline="") as myfile:
        wr=csv.writer(myfile)
        wr.writerow(["title","location","distance","description","verdict","rating","experience"])
        wr.writerows(exported)
    
       
```

```python
def parse_all():
    pages=get_all_pages()
    for page in pages:
        print(f"On scrape{page}")
        parse(url=page)
```
